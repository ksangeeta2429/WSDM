```python
import numpy as np
import codecs
import math
import pandas as pd
from googletrans import Translator
import goslate
from langdetect import detect
```

```python
songs = pd.read_csv('../Data/songs.csv', usecols = ['artist_name', 'composer', 'lyricist'])
songs['composer'] = songs['composer'].astype(str)
songs['lyricist'] = songs['lyricist'].astype(str)
```

```python
translation = []
ts = Translator() 
```

```python
df = pd.DataFrame(songs)
#print df
```

```python
headers = ['artist_name', 'composer', 'lyricist']
def translate_row(row):
    a = ts.translate(row['artist_name'], dest='en')
    b = ts.translate(row['composer'], dest='en')
    c = ts.translate(row['lyricist'], dest='en')
    return pd.Series([a.text, b.text, c.text], headers)

for i in range(100):
    df.loc[i] = translate_row(df.loc[i])

print df
```

```python
#artists = pd.Series(songs.artist_name.sort_values(inplace=False).unique())
#print len(artists) #
df['alen'] = songs['artist_name'].apply(len)
#print songs['alen'].max() #35010

#composers = pd.Series(songs.composer.sort_values(inplace=False).unique())
#print len(composers) #
df['clen'] = songs['composer'].map(str).apply(len)
#print songs['clen'].max() #7117

#lyricists = pd.Series(songs.lyricist.sort_values(inplace=False).unique())
#print len(lyricists) #33900
df['llen'] = songs['lyricist'].apply(len)
#print songs['llen'].max() #410
```
